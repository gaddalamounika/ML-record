```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
print("intercept",model.intercept_)
print("coeffient:",model.coef_)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train, y_train)
print("intercept:", model.intercept_)
print("coef:", model.coef_)
```

```python
# Step 1: Predict on the training set
y_train_pred = model.predict(X_train)

# Step 2: Calculate the error
error = y_train_pred - y_train

# Step 3: Display the first few values
print("Predicted Prices:", y_train_pred[:5])
print("Actual Prices:", y_train[:5].values)
print("Error (y_pred - y_train):", error[:5].values)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Predict on training set
y_train_pred = model.predict(X_train)

# Step 2: Calculate residuals
residuals = y_train_pred - y_train

# Step 3: Create Residual Plot using Seaborn
plt.figure(figsize=(8, 5))  # Set figure size
sns.scatterplot(x=y_train_pred, y=residuals, color='blue', alpha=0.6)

# Step 4: Add a horizontal reference line at y=0
plt.axhline(y=0, color='red', linestyle='--')

# Step 5: Customize labels and title
plt.xlabel("Predicted Prices (y_pred)")
plt.ylabel("Residuals (Error)")
plt.title("Residual Plot (Seaborn)")

# Step 6: Show the plot
plt.show()
```

```python
# Step 1: Predict on the training set
y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Predict on training set
y_train_pred = model.predict(X_train)

# Step 2: Calculate MSE
mse = mean_squared_error(y_train, y_train_pred)

# Step 3: Calculate RÂ² Score
r2 = r2_score(y_train, y_train_pred)

# Step 4: Print results
print("Mean Squared Error (MSE):", mse)
print("R Squared Value (RÂ² Score):", r2)
```

```python
y_test_pred = model.predict(X_test)
test_errors = y_test_pred - y_test

plt.scatter(y_test_pred, test_errors)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals (Errors)')
plt.title('Residual Plot - Test')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# For Train Set
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# For Test Set
mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

print(f"Train MSE: {mse_train:.2f}")
print(f"Train RÂ²: {r2_train:.2f}")
print(f"Test MSE: {mse_test:.2f}")
print(f"Test RÂ²: {r2_test:.2f}")
```

```python
import statsmodels.api as sm

# Add a constant to X_train
X_train_sm = sm.add_constant(X_train)

# Now build the model
model = sm.OLS(y_train, X_train_sm).fit()

# Print Adjusted RÂ²
print("Adjusted RÂ²:", model.rsquared_adj)
```
